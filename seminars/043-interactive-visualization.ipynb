## 2-D interactive plots

```python
%matplotlib widget    # %matplotlib inline
```

```python
import numpy as np
import pylab
import matplotlib.pyplot as plt

pylab.rcParams['figure.figsize'] = (4.3, 3)    # default figure size
```

```python
plt.figure()
x = np.linspace(-3,3,100)
y = x**3
plt.plot(x,y)
plt.grid()
```

Interactive 2-D plots offer:
- saving figure
- zoom in / zoom out
- change viewpoint

But they store data in memory. Inefficient for big data.

## 3-D plots

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2+Y**2

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
ax.view_init(25,45)   # optinally can set initial viewpoint (elev-horizontal angle, azim-vertical angle, roll-rotation around viewing direction)
plt.show()
```

```python
N=300
sigma = 0.5
T = np.linspace(0,4*np.pi,N)

X = T*np.sin(T) + sigma*np.random.randn(N)
Y = T*np.cos(T) + sigma*np.random.randn(N)
Z = T + sigma*np.random.randn(N)

X.shape, Y.shape, Z.shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')


COLORS = np.array(N//2*['r']+N//2*['b'])

ax.scatter(X, Y, Z, c=COLORS)
plt.show()
```
